```python
import os
import numpy as np
import pandas as pd
from skimage import io
from sklearn.decomposition import PCA, KernelPCA
import matplotlib.pyplot as plt
```

```python
n = 20

def norm(img):
    return (img - np.mean(img))/np.std(img)

train = pd.read_csv("../input/train.csv")
test = pd.read_csv("../input/test.csv")
data_dir = "../input/captcha_named/"
X_train, y_train = [], []
X_test, y_test = [], []

for i, f_name in enumerate(train.filename.values):
    img = io.imread(data_dir+f_name, as_gray=True)
    X_train.append(img.flatten())
    y_train.append(train.loc[i, "target"])
    
for i, f_name in enumerate(test.filename.values):
    img = io.imread(data_dir+f_name, as_gray=True)
    X_test.append(img.flatten())
    y_test.append(test.loc[i, "target"])
    
X_train = norm(np.array(X_train))
X_test = norm(np.array(X_test))
```

```python
io.imshow(X_train[n].reshape((200, 200)))
```

```python
pca = PCA(n_components = 20, whiten=True)
```

```python
pca.fit(X_train)
```

```python
fig = plt.figure(figsize=(16, 6))
for i in range(10):
    ax = fig.add_subplot(3, 10, i + 1, xticks=[], yticks=[])
    ax.imshow(pca.components_[i].reshape((200, 200)),
              cmap=plt.cm.inferno)
```

```python
print(pca.components_.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
from sklearn.metrics.pairwise import chi2_kernel
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
svm = SVC(kernel=chi2_kernel).fit(X_train, y_train)
```

```python
svm.predict(X_test)
```

```python
kpc = KernelPCA(n_components=10, kernel="rbf")
X_txf = kpc.fit_transform(X_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier, NeighborhoodComponentsAnalysis
```

```python
n_neighbors = 2
n_components= 2
nca = NeighborhoodComponentsAnalysis(n_components, random_state=11)
# Use a nearest neighbor classifier to evaluate the methods
knn = KNeighborsClassifier(n_neighbors=n_neighbors)
nca.fit(X_train, y_train)
knn.fit(nca.transform(X_train), y_train)
print(knn.score(nca.transform(X_test), y_test))
```
